```python
import pandas as pd
import numpy as np

from scipy import stats, integrate
import matplotlib.pyplot as plt
import seaborn as sns


data = pd.read_csv("/Users/chenjenny/Documents/Tyson/RD/ChickenFry.csv")
data.head()

```
